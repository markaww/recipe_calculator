```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
# Website has 56 pages of recipes so we must make 56 URLs to scrape from

tin_pages_urlist = ['https://www.recipetineats.com/recipes/']
for i in list(range(1,57)):
    tin_pages_urlist.append(f'https://www.recipetineats.com/recipes/?fwp_paged={i}')
```

```python
# Instead of a soup, we need to make a list of soups, which will be len = 56

htmls = [requests.get(url).content for url in tin_pages_urlist]
soups = [BeautifulSoup(html, 'html.parser') for html in htmls]
```

```python
# get the URL for each recipe applying the pattern to each soup

pattern = 'https:/.+?tabindex'
recipe_names_raw = [re.findall(pattern, str([i])) for soup in soups for i in soup.find_all('article')]
recipe_names_urls = [item[:-10] for sublist in recipe_names_raw for item in sublist]
```

```python
# Get category for each of the recipes compiled in the previous link

cat_htmls = [requests.get(url).content for url in recipe_names_urls]
cat_soups = [BeautifulSoup(html, 'html.parser') for html in cat_htmls]

```
